##In this assignment students have to compress racoon grey scale image into
5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.
The raw image is available in spicy.misc package with the name face.

```python
import numpy as np
from sklearn import cluster,datasets
import scipy.misc 
import matplotlib.pyplot as plt
```

```python
#loading image
racoon=scipy.misc.face(gray=True)
plt.figure(figsize=(10,5))
plt.imshow(racoon,plt.cm.gray)
plt.show()
```

```python
#convert row column data into the single array
rows=racoon.shape[0]
columns=racoon.shape[1]
im=racoon.reshape(rows*columns,1)
im.shape
```

```python
#apply k means clustering with 5 clusters
from sklearn.cluster import KMeans
model=KMeans(n_clusters=5)
model.fit(im)
```

```python
labels=model.labels_
```

```python
labels
```

```python
new_im=labels.reshape(rows,columns)
new_im
```

```python
#save the image
plt.imsave('compressed-image.png',new_im)
```

```python
#visualise new image
compres_img=plt.imread('compressed-image.png')
plt.imshow(compres_img,plt.cm.gray)
```
